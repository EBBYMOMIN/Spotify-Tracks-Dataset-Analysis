CONVERT 'DURATION' FROM MILLISECONDS TO SECONDS

```python
data.head(2)
```

```python
data['duration'] = data['duration_ms'].apply(lambda x: round(x/1000))
data.drop('duration_ms',inplace=True,axis=1)

 #Using apply() function 
 # x/1000 converts ms to sec
 # round() converts the value to nearest whole number
 
```

```python
data.duration.head()
```

** **

SETTING INDEX AS RELEASE DATE

```python
# Convert the index to datetime, specifying the correct format
data.index = pd.to_datetime(data.index, format='%d-%m-%Y', errors='coerce')

# Display the first 2 rows of the DataFrame
data.head(2)
```

** **

CREATING CORRELATION MAP.

```python
# unwanted columns are dropped
# "pearson" correlation method is applied
# set figure size to 14,6
# using seaborn heatmap function to create correlation map
# setting annotation = True, this will write the data value in each cell.
# set fmt = .1g (this is a string formating tool to use when adding annotation)
# vmin and vmax are the values to add for color map
# cmap - colormap
```

```python
# Drop the specified columns
data_subset = data.drop(["key", "mode", "explicit"], axis=1)

# Convert the entire DataFrame to numeric
data_numeric = data_subset.apply(pd.to_numeric, errors='coerce')

# Calculate correlation
correl = data_numeric.corr(method='pearson')

# Now, proceed with the visualization
plt.figure(figsize=(14, 6))
heatmap = sns.heatmap(correl, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap='inferno', linewidths=1, linecolor="Black")
heatmap = plt.title("Correlation HeatMap Between Variable")
```

WILL CREATE 2 REGRESSION PLOT USING THE DATA

```python
sample_data = data.sample(int(0.004*len(data)))

# use int function. 
# 0.004 multiplied by len of original data frame
```

```python
print(len(sample_data))
```

** **

PLOTTING A REGRESSION PLOT BETWEEN LOUDNESS AND ENERGY

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_data, y = "loudness", x = "energy", color ="c")
plt.title("Loudness VS Energy Correlation")
plt.show()
```

The regrression line is showing upside - When Energy increases Loudness of the song increases and vise-versa.

** **

PLOTTING A REGRESSION PLOT BETWEEN POPULARITY AND ACCOUSTICNESS.

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_data, y = 'popularity', x = 'acousticness', color = 'b')
plt.title('Popularity VS Acousticness Correlation')
plt.show()
```

The regression line is showing downside - If Accousticness of the song increases the Popularity of the song decreases and vice-versa.

** **

CREATE COLUMN YEARS FROM RELEASE DATE

```python
data['years'] = data.index.year 

years = data['years']

# Create a new column 'years' from the index (release_date)
```

** **

DISTRIBUTION PLOT "NUMBER OF SONGS" SINCE 1922

```python
sns.displot(years, discrete=True, aspect=2, height=5,kind='hist').set(title="Number of songs per year")
```

** **

DURATION OF SONGS OVER THE YEARS (BARPLOT)

```python
data.head(2)
```

```python
# Group by 'years' and sum the duration for each year
yearly_total_duration = data.groupby('years')['duration'].sum().reset_index()

# Plot the total duration of songs over the years
fig_dims = (18, 7)  # Dimensions
fig, ax = plt.subplots(figsize=fig_dims)
sns.barplot(x='years', y='duration', data=yearly_total_duration, ax=ax, errwidth=0)
plt.title("Total Duration of Songs Over the Years")
plt.xlabel("Year")
plt.ylabel("Total Duration (ms)")
plt.xticks(rotation=90)

# Show the plot
plt.show()
```

** **

DURATION OF SONGS OVER THE YEARS (LINEPLOT)

```python
# Group by 'years' and calculate the total duration for each year
yearly_total_duration = data.groupby('years')['duration'].sum().reset_index()

# Plot the total duration of songs over the years
fig_dims = (18, 7)  # Dimensions
fig, ax = plt.subplots(figsize=fig_dims)

# Create the line plot with the grouped data
sns.lineplot(x='years', y='duration', data=yearly_total_duration, ax=ax)
plt.title("Total Duration of Songs Over the Years")
plt.xlabel("Year")
plt.ylabel("Total Duration (ms)")
plt.xticks(rotation=90)

# Show the plot
plt.show()
```
